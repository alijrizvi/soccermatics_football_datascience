### CTEs

```python
%%sql

WITH stages AS (
    SELECT DISTINCT competition_stage_name, away_team_away_team_name,  home_team_home_team_name
    FROM df
),
stadiums AS (
    SELECT DISTINCT stadium_name, away_team_away_team_name, home_team_home_team_name
    FROM df
)
SELECT *
FROM stages
INNER JOIN stadiums ON stages.home_team_home_team_name = stadiums.home_team_home_team_name;
```

```python
%%sql

SELECT competition_stage_name, SUM(home_score) AS home_goals_scored, SUM(away_score) AS away_goals_scored
FROM (
    SELECT competition_stage_name, home_score, away_score, home_team_home_team_name, away_team_away_team_name
    FROM df
    WHERE home_team_home_team_name == 'France' 
       OR away_team_away_team_name == 'France')
GROUP BY competition_stage_name;
```

```python
%%sql

WITH away_record_fra AS (
    SELECT competition_stage_name, away_team_away_team_name, SUM(home_score) AS conceded_away
    FROM df
    WHERE away_team_away_team_name == 'France'),
home_record_fra AS (
    SELECT competition_stage_name, home_team_home_team_name, SUM(away_score) AS conceded_home
    FROM df
    WHERE home_team_home_team_name == 'France')
SELECT *
FROM home_record_fra
LEFT OUTER JOIN away_record_fra ON home_record_fra.home_team_home_team_name = away_record_fra.away_team_away_team_name;
```

* Advanced WINDOW Functions like "PERCENTILE_CONT() WITHIN GROUP (ORDER BY...)" are not Supported in SQLite, so will have to do with NTILE
* "NTILE(4)" assigns values of 1, 2, 3, or 4 for each corresponding value - 1 means the value is within the 25th Percentile; 2 means it is within the 25th-50th Percentile range, and so on

```python
%%sql

SELECT 
    home_score,
    NTILE(4) OVER (PARTITION BY competition_stage_name ORDER BY home_score) AS quantile_home_score,
    away_score,
    NTILE(4) OVER (PARTITION BY competition_stage_name ORDER BY away_score) AS quantile_away_score
FROM df
ORDER BY quantile_home_score;
```

Returning all Unique Matches Played at the Same Venue:

```python
%%sql

SELECT 
    A.stadium_name,
    A.home_team_home_team_name AS home_team, 
    B.home_team_home_team_name AS away_team
FROM df A
JOIN df B 
    ON A.stadium_name = B.stadium_name
   AND A.home_team_home_team_name < B.home_team_home_team_name   -- Ensuring Uniqueness of Match-ups
ORDER BY A.stadium_name;
```

```python
%%sql

SELECT DISTINCT
    stadium_name,
    ROW_NUMBER() OVER (PARTITION BY stadium_name ORDER BY match_date) AS match_number, -- Which Number of Match was it, Played at that Specific Stadium?
    CASE WHEN home_team_home_team_name < away_team_away_team_name 
         THEN home_team_home_team_name ELSE away_team_away_team_name END AS team1,
    CASE WHEN home_team_home_team_name < away_team_away_team_name 
         THEN away_team_away_team_name ELSE home_team_home_team_name END AS team2
FROM df
ORDER BY stadium_name, team1, team2;
```

```python
%%sql

-- Very Insightful: Classifying Goals Scored by Home/Away Teams into Quantiles

WITH quantiles_home AS (
    SELECT quantile_home_goals, ROUND(AVG(home_score), 2) AS avg_home_goals
    FROM (
    SELECT home_score,
    NTILE(4) OVER (ORDER BY home_score) AS quantile_home_goals,
    away_score,
    NTILE(4) OVER (ORDER BY away_score) AS quantile_away_goals
FROM df)
GROUP BY quantile_home_goals),
quantiles_away AS (
    SELECT quantile_away_goals, ROUND(AVG(away_score), 2) AS avg_away_goals
    FROM (
    SELECT home_score,
    NTILE(4) OVER (ORDER BY home_score) AS quantile_home_goals,
    away_score,
    NTILE(4) OVER (ORDER BY away_score) AS quantile_away_goals
FROM df)
GROUP BY quantile_away_goals)
SELECT *
FROM quantiles_home h
JOIN quantiles_away a ON h.quantile_home_goals = a.quantile_away_goals;
```

```python
%%sql

-- Goals Scored in each Phase of the Competition, Classified into Respective Phase/Stage Quantiles

SELECT competition_stage_name, quantile_home_goals, ROUND(AVG(home_score), 2) AS avg_home_goals
   FROM (
   SELECT competition_stage_name, home_score,
   NTILE(4) OVER (PARTITION BY competition_stage_name ORDER BY home_score) AS quantile_home_goals,
   away_score,
   NTILE(4) OVER (PARTITION BY competition_stage_name ORDER BY away_score) AS quantile_away_goals
   FROM df)
GROUP BY competition_stage_name, quantile_home_goals;
```
